**Denoising images with wavelets**


```python
import numpy as np
import scipy.fft
import stats.mle
import matplotlib.pyplot as plt
from PIL import Image

from IPython.display import display, clear_output

path = "/home/snake91/git/ShareCode/stats/ml/London-BW.jpg"

x = Image.open(path, "r").convert('L')

matx = np.asarray(x)

matfft = scipy.fft.fft2(matx)

#for i in np.linspace(np.min(matfft), np.max(matfft), 20):
#print(i)
matfft1 = np.where(matfft > 100000, 0, matfft)

matfft1 = np.fft.ifft2(matfft1)

y = Image.fromarray(np.uint8((matfft1)*255))
display(y)


```
